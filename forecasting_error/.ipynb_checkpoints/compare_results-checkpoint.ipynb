```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import os
```

```python
files  = os.listdir("results")

final_fig = go.Figure()

for file in files:
    
    # Get the title or paramaters of the model
    tokens = file.split("_")
    n_days = tokens[1].split("d")[0]
    cap = tokens[3]
    floor = tokens[5]
    regressor = "without regressors"
    if tokens[9] == 'True.csv':
        regressor = "with regressors"
    title = f"{n_days} day window, {cap}/{floor} cap/floor, {regressor}"
    print(title)
    
    error = pd.read_csv(f"results/{file}")
    error.set_index("Unnamed: 0", inplace=True)
    
#     fig = px.line(error)
#     fig.update_layout(title=title, height=300)
#     fig.update_xaxes(title="Forecast Hours")
#     fig.update_yaxes(title="Error ($)")
#     fig.show()

#     fig = px.line(error.abs().mean(axis=1))
#     fig.update_layout(title=title, height=300)
#     fig.update_xaxes(title="Forecast Hours")
#     fig.update_yaxes(title="Absolute Error ($)")
#     fig.show()
    
    
    final_fig.add_trace(go.Scatter(x=error.index, y=error.abs().mean(axis=1), name=title))

final_fig.update_layout(title="Model Comparison", height=500)
final_fig.update_xaxes(title="Forecast Hours")
final_fig.update_yaxes(title="Mean Absolute Error ($)")
final_fig.show()
```
