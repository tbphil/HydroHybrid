```python
import pandas as pd
```

```python
data = pd.read_excel("ALEAF_Master_HydroES_V1_101223.xlsx", 
                   sheet_name="Hourly Time Series", 
                   usecols=["Energy-DA", "Regulation Up", "Regulation Down", "Spinning Reserve", "Non-spinning Reserve"],
                   skiprows=1)
data = pd.concat([data, data], ignore_index=True)
data
```

```python
dates = []

dates += [f"Jan {i}" for i in range(1, 32)]
dates += [f"Feb {i}" for i in range(1, 29)]
dates += [f"Mar {i}" for i in range(1, 32)]
dates += [f"Apr {i}" for i in range(1, 31)]
dates += [f"May {i}" for i in range(1, 32)]
dates += [f"Jun {i}" for i in range(1, 31)]
dates += [f"Jul {i}" for i in range(1, 32)]
dates += [f"Aug {i}" for i in range(1, 32)]
dates += [f"Sep {i}" for i in range(1, 31)]
dates += [f"Oct {i}" for i in range(1, 32)]
dates += [f"Nov {i}" for i in range(1, 31)]
dates += [f"Dec {i}" for i in range(1, 32)]

hours = [f"{str(i)}" for i in range(0, 168)]

Energy_DA = pd.DataFrame(index=hours, columns=dates)
Energy_DA
```

```python
Energy_DA["Jan 1"] = data["Energy-DA"].iloc[0:24*7].values

for i, date in enumerate(dates):
    vals = data["Energy-DA"].iloc[i*24:(i+1)*24 + 144].values
    Energy_DA[date] = vals
Energy_DA
```
